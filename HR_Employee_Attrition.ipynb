```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

print(df.shape)
print(df.head())
```

```python
attrition_rate = df['Attrition'].value_counts()

print(attrition_rate)

sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Attrition',
    y='Age',
    data=df
)

plt.title("Age vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Attrition',
    y='MonthlyIncome',
    data=df
)

plt.title("Salary vs Attrition")
plt.show()
```

```python
sns.countplot(
    x='OverTime',
    hue='Attrition',
    data=df
)

plt.title("Overtime Impact")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Department',
    hue='Attrition',
    data=df
)

plt.title("Department Wise Attrition")
plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    y='JobRole',
    hue='Attrition',
    data=df
)

plt.title("Job Role Wise Attrition")

plt.show()
```

```python
sns.countplot(
    x='WorkLifeBalance',
    hue='Attrition',
    data=df
)

plt.title("Work Life Balance vs Attrition")
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(15,10))

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
print("Average Age :", df['Age'].mean())

print("Average Salary :", df['MonthlyIncome'].mean())

print("Average Years At Company :",
      df['YearsAtCompany'].mean())

print("Attrition Percentage :")

print(
    round(
        df['Attrition'].value_counts(normalize=True)*100,
        2
    )
)
```
